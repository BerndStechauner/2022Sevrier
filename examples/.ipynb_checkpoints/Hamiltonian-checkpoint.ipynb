# Hamiltonian

```python
import numpy as np
from matplotlib import pyplot as plt 
```

```python
omega=1
x=np.linspace(-5,5)
px=np.linspace(-2,2)
[X, PX]=np.meshgrid(x,px)
H=0.5*(PX**2+omega**2*X**2)
plt.pcolormesh(H)
plt.contour(H, colors='w')
```

```python
omega=1
x=np.linspace(-5,5)
px=np.linspace(-2,2)
[X, PX]=np.meshgrid(x,px)
H=0.5*(PX**2-omega**2*X**2)
plt.pcolormesh(H)
plt.contour(H, colors='w')
```

```python
omega=1
x=np.linspace(-5,5)
px=np.linspace(-2,2)
[X, PX]=np.meshgrid(x,px)
H=0.5*(PX**2+omega**2*np.cos(X))
plt.pcolormesh(H)
plt.contour(H, colors='w')
```

```python
omega=1
x=np.linspace(-2,2)
px=np.linspace(-2,2)
[X, PX]=np.meshgrid(x,px)
H=0.5*PX**2-0.5*X**2+.25*X**4
plt.pcolormesh(H)
plt.contour(H,10, colors='w')
```
